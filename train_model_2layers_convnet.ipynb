```python
from keras.optimizers import Adam
SHAPE = (1024, 1024, 4)
n_out = 28
model = create_model(SHAPE, n_out)
model.compile(
    loss='binary_crossentropy',
    optimizer= Adam(1e-04),
    metrics=['acc',f1])

model.summary()
```

### Train model

```python
from Weight_for_loss import weights_for_loss
```

```python
epochs = 10; batch_size = 20
checkpointer = ModelCheckpoint(
    './working/yixiao/Convnet.model', 
    verbose=2, 
    save_best_only=True)

# split and suffle data 
np.random.seed(2018)
indexes = np.arange(train_dataset_info.shape[0])
np.random.shuffle(indexes)
train_indexes = indexes[:27500]
valid_indexes = indexes[27500:]

# create train and valid datagens
train_generator = data_generator_4channels.create_train(
    train_dataset_info[train_indexes], batch_size, (1024,1024,4), augument=True)
validation_generator = data_generator_4channels.create_train(
    train_dataset_info[valid_indexes], 100, (1024,1024,4), augument=False)

# train model
history = model.fit_generator(
    train_generator,
    steps_per_epoch=5,
    validation_data=next(validation_generator),
    epochs=epochs, 
    verbose=1,
    class_weight= weights_for_loss,
    callbacks=[checkpointer])

model.save_weights('./working/yixiao/Convnet_weights.h5')
with open('./working/yixiao/Convnet_architecture.json', 'w') as f:
    f.write(model.to_json())
```
